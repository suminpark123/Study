# 컴퓨터가 이해 할 수 있는 HTML형태로 변형

## beautifulSoup 라이브러리
### request통해서 넘겨받은 문자형태 데이터를 html데이터로 변형

```python
# 임포트할때 스톤숩은 사용하지 않는다!
from bs4 import BeautifulSoup as bs
```

```python
# bs("내가 가지고있는 텍스트 정보","파싱방법")
soup=bs(res.text,'lxml')
```

```python
# 사용자 : 화면에서 블로그,카페.지식인 등의 글자를 수집
# 컴퓨터 : 블로그,커페,지식인의 태그정보를 수집
# select("내가 가지고올 요소의 선택자")
#select_one = 복수개의 데이터에[서 한개만 가지고 온다.
# ex)배열 데이터라면 첫번째 데이터(index 0)만 가지고 온다.
data=soup.select('a.nav')
soup.select_one("a.nav")
```

```python
# 복수개의 데이터를 요청하게된다면 리턴하는 데이터는 타입이 배열(리스트)
# 인덱스 번호를 통해서 데이터를 조회

data[0].text
```

```python
# 전체데이터를 출력
for i in data :
    print(i.text)
```
