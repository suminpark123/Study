### iframe
- 화면에는 존재하지만 실제데이터는 다른공간에 존재하는 경우
- iframe 데이터 수집하는 방법
- 1.f12개발자도구로 가서 ctrl+F ->iframe
- 2.실제로 사용되고있는 iframe 찾기 (화면색이 변함)
- 3.iframe 이 가지고 있는 경로를 통해서 실제 주소로 찾아가기('src')
- 3-1.url은 메인서버주소+경로

```python
# 진짜 데이터를 가지고있는 서버의 url수집 ->요청
res3=req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
soup3=bs(res3.text,'lxml')
```

```python
review1=soup3.select('div.score_reple>p>span+span') #div.score_reple>p>span:nth-child(2)
```

```python
review1[0].text.strip()
```

```python
for i in review1:
    print(i.text.strip())
```

```python
#1. 데이터를 수집할때 선택자를 정확하게 쓰는것이 제일 좋다.
#2. 그 선택자를 사용 할 수 없는 경우가 있다면? (span 태그ㅜ가 하나만 존재)
#3. 불필요한 글자(관람객)만 삭제 (많이 사용되지는 않음)
```

```python
# 불필요한 요소인 '관람객'글자의 태그 찾아서 알려주기
viewer=soup3.select('span.ico_viewer')
```

```python
# 리스트형태의 데이터이기 때문에 하나씩 꺼내서 지워주기->extract()
for i in viewer:
    i.extract()
```

```python
for i in review1:
    print(i.text.strip())
```

### 1페이지부터 10페이지까지 정보수집하기!

```python
#코드 합치기
#반복문 돌리기-> 이 상태로는 1페이지만 10번

for i in range(10):
    res3=req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
    soup3=bs(res3.text,'lxml')

    review1=soup3.select('div.score_reple>p') 
    viewer=soup3.select('span.ico_viewer')
    for i in viewer:
        i.extract()

    for i in review1:
        print(i.text.strip())
```

```python
#코드 합치기
#반복문 돌리기 ->수정해줘야 하는 것은? 요청하는 페이지 주소

for i in tq(range(10)):
    res3=req.get(f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={i}")
    soup3=bs(res3.text,'lxml')

    review1=soup3.select('div.score_reple>p') 
    viewer=soup3.select('span.ico_viewer')
    for i in viewer:
        i.extract()

    for i in review1:
        print(i.text.strip())
```
