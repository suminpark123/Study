```python
soup=bs(res.text,'lxml')
```

# 노래제목 수집

```python
# 노래제목을 검사했을떄 a태그로만 작성이 돼있다.
# 클래스나 아이디 등 구분할 구분자가 없다
# 부모태그를 검사해라. 단 , 부모태그 또한 구분자가 없는 경우에는
# 구분자가 존재하는 부모태그까지 검사
# ★클래스가 중간에 공백이 있는 경우에는 클래스값이 복수개
# 해결책 : 클래스값들을 .클래스.클래스로 공백을 지우고 이어준다.
song=soup.select('div.ellipsis.rank01>span>a')
```

# 가수 이름 수집

```python
# 가수명을 검사해서 태그정보를 알아온 다음에
# select 를 통해서 데이터를 수집
```

```python
#개수가 정해져있는 데이터를 수집할때는 항상 데이터 개수를 파악
```

```python
singer=soup.select('span.checkEllipsis')
```

```python
len(song)
```

```python
len(singer)
```

```python
for i in singer :
    print(i.text)
```

```python
for i in song :
    print(i.text)
```

```python
# a태그를 따로 들고오는게 아니고 ,하나로 묶어서 가지고 온다.
# 묶음으로 들고 왔을 경우에, 가수명이 중복이 된다.
# 그 이유는 ,밑에 숨어있던 span태그까지 들고 왔기 때문에
singer=soup.select('div.ellipsis.rank02>span.checkEllipsis')
```

```python
len(singer)
```

```python
for i in singer : 
    print(i.text)
```

```python
# 수집한 데이터를 리스트로 묶어주기
song_list=[]
singer_list=[]
rank_list=[]

for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
song_list
```

```python
# 수집한 데이터 DataFrame으로 만들어주기 -dictionary 구조활용

dic={'노래제목':song_list,'가수':singer_list,'순위':rank_list}
```

```python
melon=pd.DataFrame(dic)
```

```python
melon.set_index('순위',inplace=True)
```

```python
melon.to_csv('멜론차트.csv',encoding='CP949')
```
