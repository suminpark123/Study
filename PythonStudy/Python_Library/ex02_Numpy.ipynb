# 최종 실습
 - 영화 평점 리뷰 데이터 분석

```python
# 가져온 데이터 파일 읽어오기(불러오기)
# numpy.loadtxt('파일이름.확장자까지!',delimiter="해당 파일의 구분기호", dtype="변환할 타입")

data=np.loadtxt('ratings.txt',delimiter="::", dtype="int")
data
```

```python
# 배열의 차원을 구하는 기능 : ndim

print(f"배열의 차원 : {data.ndim}")
# 배열의 크기를 구하는 기능 : shape

print(f"배열의 크기 : {data.shape}")
# 배열의 개수를 구하는 기능 : size

print(f"배열의 개수 : {data.size}")
```

```python
print("*"*5)
```

```python
# 전체 평점(별 개수)의 평균 구하기 -->mean()

r_mean=data[:,2].mean()
```

```python
# numpy에서 제공하는 unique() 함수
# 해당하는 데이터 중 중복된 값들을 전부 제거한 뒤 고유 데이터만 추출

user_id=np.unique(data[:,0])

print(user_id)
print(user_id.shape)
```

```python
# 사용자 아이디만 boolean indexing

#사용자 아이디가 1인 데이터만 boolean indexing
a_data=data[data[:,0]]
a_data[a_data==1]

mv=[]

for i in user_id:
    data_user=data[data[:,0]==i,:]
    value=data_user[:,2].mean()
    mv.append([i,value])
    
print("결과 :",mv[:])


```

```python
li=[]
for i in range(1,6):
    user_mean=data[data[:,0]== i, 2].mean()
    li.append([i,user_mean])
li
```

```python

user_rating=[]

#1. 각 사용자별 고유 번호들을 하나하나 접근하여 그 중에서 평점이 있는 인덱스열에 접근
# 접근한 열을 기준으로 평균값 mean() 구함
for user in user_id:
    data_user_mean=data[data[:,0]==user,2].mean()
    #3. 마련된 공간에 데이터 차곡차곡 쌓아주기
    user_rating.append([user,data_user_mean])
user_rating    
```
