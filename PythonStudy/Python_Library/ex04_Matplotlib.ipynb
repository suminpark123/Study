```python
import matplotlib.pyplot as plt
```

```python
y= [2,4,6]

# plot 함수() - 일직선으로 되어있는 Lineplot 그리기
plt.plot(y) #그래프 그리기
plt.show() #그래프 출력
```

```python
x= [1,2,3]
y= [2,4,6]


plt.plot(x,y) 
plt.show() 
```

```python
#line style =ls
# 그려지는 선의 스타일을 지정하는 속성
plt.plot(x,y, ls='--')
plt.show()
```

```python
# marker
plt.plot(x,y, marker='s')
plt.show()
```

```python
#line width =lw
#선의 두께
plt.plot(x,y, linewidth=10)
plt.show()
```

```python
# line color
# 선의색상 변경하기 RGB 색상표 이용가능
plt.plot(x,y, color='#553939',marker='o',ls='--')
plt.show()
```

```python
#grid
#틱의 위치를 잘 보기 위해

plt.plot(x,y, color='r',marker='s',mfc='b',ls=':',ms='10')
plt.xlim(-2,8) #(최소,최대)
plt.ylim(-1,10)
plt.xticks([0,3,6])
plt.yticks([1,5,6])
plt.grid()


plt.show()
```

```python
# 여러 그래프 동시에 그리기
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x,color='r',label='x')
plt.plot(y,color='b',label='y')
plt.plot(z,color='orange',label='z')
plt.legend(loc='upper center')
plt.show()

#'best', 'upper right', 'upper left', 'lower left', '
#lower right', 'right', 'center left', 'center right', '
#lower center', 'upper center', 'center'
```

```python
# 한글 지원 폰트로 바꾸기
from matplotlib import rc
rc('font',family='Malgun Gothic')
```

```python
# x축,y축 이름
plt.plot(x,color='r',label='엑스')
plt.plot(y,color='b',label='와이')
plt.plot(z,color='orange',label='제트')
plt.legend(loc='upper center')
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('x,y,z 그래프')
plt.show()
```

- 어떤 요일에 가장 많은/적은 사고가 발생했을지?

```python
# 1. 요일별 사고 발생 건수 확인하기
# 1-1, '요일' 컬럼 가져오기
data['요일']
```

```python
#중복 없는 값 확인 가능
data['요일'].unique()
```

```python
# 1-2, 요일별 발생 건수 확인
acc=data['요일'].value_counts()
```

```python
# 2. 요일 순서대로 인덱싱해서 보기 편하게 만들어주자
y=acc[['월','화','수','목','금','토','일']]
```

```python
x=y.index
```

```python
plt.bar(x,y)
#lt.xticks(['월','화','수','목','금','토','일'])
#lt.yticks([0,100,200,300,400,500,600])
plt.show()
```

```python

plt.bar(x,y)
plt.ylim(500,620)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 사망교통사고 건수')
plt.show()
```
