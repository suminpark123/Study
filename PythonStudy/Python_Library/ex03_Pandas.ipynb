```python
df5
```

```python
pd.merge(df5,df6,on='key')
```

```python
# how -병합의 방식 지정
# (concat의 join)
# outer : 전체 값 출력
# inner : 공통된 데이터의 값만 출력
# left :왼쪽 데이터(먼저 적은 데이터를 기준으로 출력)
# right : 오른쪽 데이터 (나중에 적은 데이터를 기준으로 출력)
```

```python
pd.merge(df5,df6,on='key',how='left')
```

```python
df7 =pd.merge(df5,df6,on='key',how='right')
```

```python
# 결측치 채우기 -fillna
# inplace 를 true 로 하면 반영
df7.fillna(0, inplace=True)
```
