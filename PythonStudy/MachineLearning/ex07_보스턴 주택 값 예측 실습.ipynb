### Ridge vs Lasso
- Lasso : L1규제, 특정 계수들(w)이 0이 될 수 있음(자동 특성 선택)
- Ridge : L2규제, 골고루 규제 적용, 0에 가까워짐 (0이 되지 않음)

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
alpha_list = [0.001,0.01,0.1,1,10,100,1000]
ridge_coef_list = []  # 가중치 리스트 /계수값 담으르 리스트
lasso_coef_list = [] 

for a in alpha_list:
    r_model = Ridge(alpha=a)  #규제 강도에 따른 모델 생성
    l_model = Lasso(alpha=a)
    
    r_model.fit(X_train,y_train) #모델 학습 -> 계수들을 찾음
    l_model.fit(X_train,y_train)
    
    ridge_coef_list.append(r_model.coef_)
    lasso_coef_list.append(l_model.coef_)
```

```python
ridge_coef_list
```

```python
X_train.head()
```

```python
ridge_df=pd.DataFrame(np.array(ridge_coef_list).T,
            columns=alpha_list)  # 전치
```

```python
lasso_df=pd.DataFrame(np.array(lasso_coef_list).T,
            columns=alpha_list)  # 전치
```
